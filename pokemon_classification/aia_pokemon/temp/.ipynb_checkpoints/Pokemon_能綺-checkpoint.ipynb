```python
features['appearedtime'] = features['appearedHour']+(features['appearedMinute']/60)
maxtime=features['appearedtime'].max()
```

```python
features['appearedtime_sin'] = np.sin(2 * np.pi * features['appearedtime']/maxtime)
features['appearedtime_cos'] = np.cos(2 * np.pi * features['appearedtime']/maxtime)
```

```python
features['temperature']=(features['temperature']-features['temperature'].min())/(features['temperature'].max()-features['temperature'].min())
features['windSpeed']=(features['windSpeed']-features['windSpeed'].min())/(features['windSpeed'].max()-features['windSpeed'].min())
features['pressure']=(features['pressure']-features['pressure'].min())/(features['pressure'].max()-features['pressure'].min())
features['population_density']=(features['population_density']-features['population_density'].min())/(features['population_density'].max()-features['population_density'].min())
features['gymDistanceKm']=(features['gymDistanceKm']-features['gymDistanceKm'].min())/(features['gymDistanceKm'].max()-features['gymDistanceKm'].min())
features['pokestopDistanceKm']=(features['pokestopDistanceKm']-features['pokestopDistanceKm'].min())/(features['pokestopDistanceKm'].max()-features['pokestopDistanceKm'].min())
features['appearedtime_sin']=(features['appearedtime_sin']-features['appearedtime_sin'].min())/(features['appearedtime_sin'].max()-features['appearedtime_sin'].min())
features['appearedtime_cos']=(features['appearedtime_cos']-features['appearedtime_cos'].min())/(features['appearedtime_cos'].max()-features['appearedtime_sin'].min())
```

```python
batch_size = 400
epochs = 200
lr = 0.01
train_ratio = 0.95
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y_one_hot, 
                                                    test_size=0.05, 
                                                    stratify=y,random_state=42)
```

```python
from keras.datasets import mnist
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from keras.models import Sequential
from keras.utils import np_utils
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
```

```python
model =Sequential()
model.add(Dense(88, activation='relu', input_shape=(298,)))
model.add(Dropout(0.6))
model.add(Dense(66, activation='relu')) 
model.add(Dropout(0.6))
model.add(Dense(6, activation='softmax')) 
print(model.summary())


  # model.add(Dropout(0.2, input_shape=(60,)))
   # model.add(Dense(60, init='normal', activation='relu', W_constraint=maxnorm(3))) model.add(Dense(30, init='normal', activation='relu', W_constraint=maxnorm(3))) model.add(Dense(1, init='normal', activation='sigmoid'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer="Nadam",
              metrics=['accuracy'])
```

```python
#early_stopping = EarlyStopping(monitor='val_loss', patience=50, verbose=2)

filepath="weights-best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=2, save_best_only=True, save_weights_only=False)

model_history = model.fit(x=X_train, y=y_train,
                          batch_size=batch_size,
                          epochs=epochs,
                          validation_split=1 - train_ratio, callbacks=[checkpoint])
```

```python
train_loss = model_history.history['loss']
train_acc = model_history.history['acc']
valid_loss = model_history.history['val_loss']
valid_acc = model_history.history['val_acc']
```

```python
plt.plot(train_loss, 'b', label='train')
plt.plot(valid_loss, 'r', label='valid')
plt.legend()
plt.title("Loss")
plt.show()

plt.plot(train_acc, 'b', label='train')
plt.plot(valid_acc, 'r', label='valid')
plt.legend(loc=4)
plt.title("Accuracy")
plt.show()
```

```python
model_best= load_model(filepath)
```

```python
y_predict = model_best.predict_classes(X_sub)
```
